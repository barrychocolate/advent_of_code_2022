```python
import pandas as pd
import numpy as np
```

```python
# Test Data
# A Y
# B X
# C Z

d = {'elf': ['A', 'B', 'C'], 'me': ['Y', 'X', 'Z']}
data = pd.DataFrame(data=d)
```

```python
# Actual data
colspecs = [(0, 1), (1, 2)]
data = pd.read_fwf("day_2.txt", names=['elf', 'me'])
data
```

```python
# calculate the scores for the shapes
data['elf_score']= data['elf'].apply(lambda x: 1 if x=='A' else 2 if x=='B' else 3)
data['me_score']= data['me'].apply(lambda x: 1 if x=='X' else 2 if x=='Y' else 3)
data['outcome']= np.nan # add outcome column
data
```

```python
# calculate score for the round
def outcome(them, me):
    if me - them in (1, -2):
        #win
        return me + 6
    elif me-them == 0:
        #draw
        return me + 3
    else:
        #lost
        return me
```

```python
# test all outcomes
# print(outcome(1, 1)) # 4
# print(outcome(1, 2)) # 8
# print(outcome(1, 3)) # 3
# print(outcome(2, 1)) # 1
# print(outcome(2, 2)) # 5
# print(outcome(2, 3)) # 9
# print(outcome(3, 1)) # 7
# print(outcome(3, 2)) # 2
# print(outcome(3, 3)) # 6
```

```python
# Calculate the outcomes
for i, row in data.iterrows():
    data.at[i,'outcome'] = outcome(row['elf_score'], row['me_score'])
    
data
```

```python
P1 = data['outcome'].sum()

print(f'The answer to part 1 is {P1}')
```

# Part 2

```python
# X means you need to lose
# Y means you need to end the round in a draw
# Z means you need to win. Good luck!"

# Update me_score based on the new understanding of the me column
for i, row in data.iterrows():
    if row['me'] == 'Y':
        #draw
        data.at[i,'me_score'] = row['elf_score']
    elif row['me'] == 'X':
        # need to lost
        if row['elf_score'] == 1:
            data.at[i,'me_score'] = 3
        elif row['elf_score'] == 2:
            data.at[i,'me_score'] = 1
        else:
            data.at[i,'me_score'] = 2
    else:
        # need to win
        if row['elf_score'] == 1:
            data.at[i,'me_score'] = 2
        elif row['elf_score'] == 2:
            data.at[i,'me_score'] = 3
        else:
            data.at[i,'me_score'] = 1

data
```

```python
# Calculate the outcomes
for i, row in data.iterrows():
    data.at[i,'outcome'] = outcome(row['elf_score'], row['me_score'])
    
data
```

```python
P2 = data['outcome'].sum()

print(f'The answer to part 2 is {P2}')
```
